```python
#copy to temper
table_input = table_input_orig.copy()

#clean summoner names as in riot's systems
table_input.index = table_input.index.str.lower()
table_input.index = table_input.index.str.replace(' ', '')
print table_input.head()
print len(table_input.index)
```

```python
i = 0
x = 0
list_40 = []

list_index_ids = []
data_aux1_id = []
data_aux1_summoner = []
data_aux1_rango =[]

data_error_list =[]
    
while i < (len(list_index)/40 + 1):  
    list_40 = list_index[40*i:40*(i+1)]
    #get all summoner ids
    #WARNING: here, we need to do requests of <= 40 summoner names
    r = rq.get('https://euw.api.pvp.net/api/lol/euw/v1.4/summoner/by-name/' +str(list_40).strip('[]').replace("'", "")+'?api_key=' + apikey)
    #check response code
    print r.status_code
    list_ids = json.loads(r.text)
    #add all id's to the list
    for k, v in list_ids.iteritems():
        if (str(k.encode("utf-8")) not in table_input.index):
            print k,v['id']
            data_error_list.append([k, v['id']])
        else:
            print k,v['id'],table_input.loc[str(k), 'Rango']
            data_aux1_id.append(v['id'])
            data_aux1_summoner.append(k)
            data_aux1_rango.append(table_input.loc[str(k), 'Rango'])
            list_index_ids.append(v['id'])
    i = i + 1
    time.sleep(2)
```

```python
i = 0
list_10 = []

data_aux2_id = []
data_aux2_rank = []
    
while i < (len(list_index_ids)/10 + 1):  
    list_10 = list_index_ids[10*i:10*(i+1)]
    
    #request all summoner ranks
    #WARNING: here, requests are limited to <= 10 summoner id's at once
    r = rq.get('https://euw.api.pvp.net/api/lol/euw/v2.5/league/by-summoner/' +str(list_10).strip('[]')+'?api_key=' + apikey)

    #check response code
    print r.status_code

    list_index_rank = json.loads(r.text)

    for k, v in list_index_rank.iteritems():
        #print k, v[0]['tier']
        data_aux2_id.append(int(k))
        data_aux2_rank.append(v[0]['tier'])
    i = i + 1
    time.sleep(2)
```
