```python
teams_no_dupes['Event'] = teams_no_dupes['League']+teams_no_dupes['Year'].map(str)+teams_no_dupes['Season']+teams_no_dupes['Type']
teams_no_dupes.head()
```

```python
G_event=nx.Graph()
G_players=nx.Graph()

for index, row in teams_no_dupes.iterrows():
    
    G_event.add_node(row['Event'], bipartite = 0)
    
    teammates = [row['Top'], row['Jungle'], row['Middle'], row['ADC'], row['Support']]
    
    G_event.add_nodes_from(teammates, bipartite=1)
    
    G_event.add_edges_from([(row['Event'], row['Top']), (row['Event'], row['Jungle']), (row['Event'], row['Middle']), 
                      (row['Event'], row['ADC']), (row['Event'], row['Support'])])
    

for index, row in teams.iterrows():
    
    teammates = [row['Top'], row['Jungle'], row['Middle'], row['ADC'], row['Support']]
    
    for a, b in itertools.combinations(teammates, 2):
        if G_players.has_edge(a,b):
            G_players[a][b]['weight'] += 1 
        else:
            G_players.add_edge(a,b, weight = 1)
```

```python
nx.is_bipartite(G_event)
```

```python
len(list(nx.connected_components(G_event)))
```

```python
graphs = list(nx.connected_component_subgraphs(G_event))
```

```python
top_nodes = {n for n, d in G_event.nodes(data=True) if d['bipartite']==0}
bottom_nodes = set(G_event) - top_nodes

H = bipartite.projected_graph(G_event, bottom_nodes)
H_w = bipartite.weighted_projected_graph(G_event, bottom_nodes)
```

```python
nx.write_graphml(G_event, "LOL_ProNet_Event.graphml")
nx.write_graphml(H, "LOL_ProNet_Event_Projected.graphml")
nx.write_graphml(H_w, "LOL_ProNet_Event_Projected_W.graphml")

nx.write_graphml(G_players, "LOL_ProNet_Player.graphml")
```
