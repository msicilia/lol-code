```python
G_players_competitive=nx.Graph() 

for index, row in teams_fix.iterrows():
    
    teammates_red = [row['redTop'], row['redJungle'], row['redMiddle'], row['redADC'], row['redSupport']]
    teammates_blue = [row['blueTop'], row['blueJungle'], row['blueMiddle'], row['blueADC'], row['blueSupport']]
    
    for a, b in itertools.combinations(teammates_red, 2):
        if G_players_competitive.has_edge(a,b):
            G_players_competitive[a][b]['weight'] += 1 
            G_players_competitive[a][b]['teammate'] = True
        else:
            G_players_competitive.add_edge(a,b, weight = 1, teammate= True, opponent = False)
    
    for a, b in itertools.combinations(teammates_blue, 2):
        if G_players_competitive.has_edge(a,b):
            G_players_competitive[a][b]['weight'] += 1
            G_players_competitive[a][b]['teammate'] = True
        else:
            G_players_competitive.add_edge(a,b, weight = 1, teammate= True, opponent = False)
            
    for a, b in itertools.product(teammates_red, teammates_blue):
        if G_players_competitive.has_edge(a,b):
            G_players_competitive[a][b]['weight'] += 1
            G_players_competitive[a][b]['opponent'] = True
        else:
            G_players_competitive.add_edge(a,b, weight = 1, teammate = False, opponent= True)
            
nx.write_graphml(G_players_competitive, "LOL_ProNet_Player_Competitive.graphml")
```
